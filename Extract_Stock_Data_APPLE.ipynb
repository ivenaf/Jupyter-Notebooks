### 2. Extract Stock Data with yfinance Library

```python
# Ticker() module to be used. We create an object that allows to access functions to extract data. 
 
# Ticker symbol must be known beforehand. Example: Apple

apple = yf.Ticker("AAPL")
```

In yfinance, the Ticker class is used to create an object that holds all the information about a specific stock. This object acts as your interface to access various data and functionalities related to that stock.

Now we have a Python object named 'apple' and can access function and variables to extract the type of data we need (historical market data, financial information, company informatino, stocks splits, dividends, real-time data.

### 3. Get/Create a JSON File from the data extraction and download

```python
import yfinance as yf
import json
```

```python
apple_data = {
    "info": apple.info,
    "history": apple.history(period='max').to_json(),
    "financials": apple.financials.to_json()
}
```

```python
# save data to json
with open('apple_data.json','w') as f:
    json.dump(apple_data, f, indent=4)

print("Apple saved as apple_data.json")
```

### 4. Stock Info - access only specifics like info, country etc. 

```python
with open('apple_data.json') as json_file:
    apple_info = json.load(json_file)
    apple_info = apple_data['info'] # to access only the 'info' key

apple_info

# you could also get the info beforehand without downloading the json file 
# by simple apple.info and the yf library gets it for you
```

### 5. Extracting Share price

```python
#Using the period parameter we can set how far back from the present to get data. 
#The options for period are 1 day (1d), 5d, 1 month (1mo) , 3mo, 6mo, 1 year 
#(1y), 2y, 5y, 10y, ytd, and max.
apple_share_price_data = apple.history(period="max")
apple_share_price_data.head()

# the format returned in is Pandas DF. 
# Date as index
```

```python
# Reset index of the dataframe , currently it is the Data
# inplace parameter True so changes takes place in dataframe itself

apple_share_price_data.reset_index(inplace=True)
```

```python
apple_share_price_data
```

### 6. Plotting a Line Chart for Opening Price Development and 200 Day MA

```python
import matplotlib.pyplot as plt
```

```python
# plot the OPEN price against the Date:
plt.figure(figsize=(25,8))

apple_share_price_data.plot(x='Date',y='Open', color = 'red', label = 'Opening Price')



# add a 200MA to smooth out short-term flcutuations and higlight long-term trend
apple_share_price_data['MA 200'] = apple_share_price_data['Open'].rolling(window=200).mean()
plt.plot(apple_share_price_data.index, apple_share_price_data['MA 200'], color = 'blue', label = '200 Day MA') 

plt.title("Opening Share Price 1980-2025")
plt.xlabel("Date")
plt.ylabel("Opening Price USD")
plt.legend()
plt.grid(True, alpha=0.3)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
```
