#### Importing dataset, first overview and initial cleansing

```python
df = pd.read_csv('world-happiness-report-2021.csv')
df.head()
```

```python
df.columns=  ['Country_name', 'Regional_indicator', 'Ladder_score',
       'Standard_error_of_ladder_score', 'upperwhisker', 'lowerwhisker',
       'Logged_GDP_per_capita', 'Social_support', 'Healthy_life_expectancy',
       'Freedom_to_make_life_choices', 'Generosity',
       'Perceptions_of_corruption', 'Ladder_score_in_Dystopia',
       'Explained_by:_Log_GDP_per_capita', 'Explained_by:_Social_support',
       'Explained_by:_Healthy_life_expectancy',
       'Explained_by:_Freedom_to_make_life_choices',
       'Explained_by:_Generosity', 'Explained_by:_Perceptions_of_corruption',
       'Dystopia_+_residual']
```

```python
# top 15 ladder score
df_tops=df.loc[df['Ladder_score']>7]

print('Shape', df_tops.shape)
top15=df_tops.head(15)
top15
```

#### Histogram, Barchart Distribution of Ladder Score

```python
plt.figure(figsize=(12,3))
plt.hist(df.Ladder_score, bins=14, color=['#f27750'], label= ['Ladder_score'], rwidth=0.7)
plt.title('Histogram Ladder Score')
plt.xlabel('Ladder Score')
plt.ylabel('Frequency')
plt.legend();
```

#### Histogram 2d - Ladder Score and Social Support

```python
x=df.Ladder_score
y=df.Social_support

plt.figure(figsize=(6,3))
plt.hist2d(x, y, bins= 5, cmap='Blues')
plt.title('Histogram Ladder Score and Social Support')
plt.xlabel('Values Ladder Score')
plt.ylabel('Values Social Support')
plt.colorbar();
```

#### Distribution Ladder Score and Dystopia

```python

sns.displot(x='Ladder_score', data=df, kde=True);
sns.displot(x='Dystopia_+_residual', data=df, kde=True, color='red');
plt.show()
```

#### Top/Bottom 15 scores

```python
plt.figure(figsize=(8,2))
sns.countplot(y='Regional_indicator', data=top15, color='green')
plt.title('Highest ladder score')
plt.show()
plt.figure(figsize=(8,2))
sns.countplot(y='Regional_indicator', data=bottom15)
plt.title('Lowest ladder score')
plt.show()
```

```python
# df with only numerical values
df_num = df.drop(['Country_name', 'Regional_indicator'], axis=1)
df_num.head()
```

#### Heatmap Correlations

```python
#correlation coefficient between variables
# remove columns with 'Explained by' in the name and remove upperwhisker and lowerwhisker
df_num = df_num.loc[:,~df_num.columns.str.contains('Explained_by')]
#df_num = df_num.drop(['upperwhisker', 'lowerwhisker'], axis=1, inplace=False) 
#df_num = df_num.drop(['Ladder_score_in_Dystopia'], axis=1, inplace=False)
df_num = df_num.drop(['Standard_error_of_ladder_score'], axis=1, inplace=False)

df_num.head()

plt.figure(figsize=(11,7))
sns.heatmap(df_num.corr(), annot=True, cmap='Spectral')
plt.title('Correlation coefficient between variables')

plt.show()
```
